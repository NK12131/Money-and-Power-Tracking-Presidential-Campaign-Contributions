```python
import pandas as pd

# ------------------------------------------------------------------
# 1. Load committee master
# ------------------------------------------------------------------
comm_path = "/Users/nithinkumar/Desktop/untitled folder/Committee/committees_cleaned_1.csv"
comm = pd.read_csv(comm_path)

print("Original committee_master shape:", comm.shape)
print("Columns:", list(comm.columns))

# ------------------------------------------------------------------
# 2. Canonical Trump candidate + committees (REAL ones)
# ------------------------------------------------------------------
TRUMP_CAND_ID = "P80001571"
TRUMP_PARTY   = "REPUBLICAN PARTY"

TRUMP_COMMITTEES = {
    "C00580100": "DONALD J. TRUMP FOR PRESIDENT, INC.",
    "C00618389": "TRUMP MAKE AMERICA GREAT AGAIN COMMITTEE",
    "C00767061": "SAVE AMERICA JOINT FUNDRAISING COMMITTEE",
    "C00839051": "TRUMP NATIONAL COMMITTEE JFC",
    "C00823870": "MAGA INC",
    "C00596909": "TRUMP VICTORY",
    "C00694323": "DONALD J. TRUMP FOR PRESIDENT 2024, INC."
}

# ------------------------------------------------------------------
# 3. Map your actual column names (handles upper/lower/space/underscore)
# ------------------------------------------------------------------
colmap = {c.lower().replace(" ", "_"): c for c in comm.columns}

cid_col   = colmap.get("committee_id")         # "committee_id" or "Committee Id"
cand_col  = colmap.get("candidate_ids") or colmap.get("candidate_id")
name_col  = colmap.get("name")
party_col = colmap.get("party_full")
desig_col = colmap.get("designation_full")
ctype_col = colmap.get("committee_type_full") or colmap.get("committee_type")
state_col = colmap.get("state") or colmap.get("state1")

print("\nResolved column mapping:")
print("committee_id  ->", cid_col)
print("candidate_ids ->", cand_col)
print("name          ->", name_col)
print("party_full    ->", party_col)
print("designation   ->", desig_col)
print("committee_type->", ctype_col)

if cid_col is None or cand_col is None:
    raise ValueError("Could not resolve committee_id or candidate_ids column names  check comm.columns")

# ------------------------------------------------------------------
# 4. Update existing rows OR add new ones for each Trump committee
# ------------------------------------------------------------------
new_rows = []
updated = 0

for cmte_id, cmte_name in TRUMP_COMMITTEES.items():
    mask = comm[cid_col].astype(str) == cmte_id

    if mask.any():
        # Update existing row
        comm.loc[mask, cand_col] = TRUMP_CAND_ID
        if party_col:  comm.loc[mask, party_col] = TRUMP_PARTY
        if desig_col:  comm.loc[mask, desig_col] = "Authorized by a candidate"
        # Very rough default types  you can refine per cmte_id if needed
        if ctype_col:  comm.loc[mask, ctype_col] = "Presidential"
        updated += mask.sum()
    else:
        # Create a new row
        row = {
            cid_col: cmte_id,
            cand_col: TRUMP_CAND_ID,
        }
        if name_col:  row[name_col]  = cmte_name
        if party_col: row[party_col] = TRUMP_PARTY
        if desig_col: row[desig_col] = "Authorized by a candidate"
        if ctype_col: row[ctype_col] = "Presidential"
        if state_col: row[state_col] = "UNKNOWN"

        new_rows.append(row)

# Append new rows if any
if new_rows:
    comm = pd.concat([comm, pd.DataFrame(new_rows)], ignore_index=True)

print(f"\nTrump committees updated: {updated}")
print(f"Trump committees newly added: {len(new_rows)}")
print("New committee_master shape:", comm.shape)

# Quick sanity check  all rows with Trump's candidate ID
trump_comm = comm[comm[cand_col] == TRUMP_CAND_ID]
print("\nRows in committee_master with Trump candidate ID:")
print(trump_comm[[cid_col, name_col, cand_col]].drop_duplicates().head(20))

```

```python
# --------------------------------------------------------------
# PATCH BIDEN COMMITTEES
# --------------------------------------------------------------

BIDEN_CAND_ID = "P80000722"   # Official Biden candidate ID from FEC
BIDEN_PARTY   = "DEMOCRATIC PARTY"

BIDEN_COMMITTEES = {
    "C00743975": "BIDEN FOR PRESIDENT (2020 PCC)",
    "C00804357": "BIDEN FOR PRESIDENT 2024 (2024 PCC)",
    "C00694323": "TEAM BIDEN-HARRIS JOINT FUNDRAISING COMMITTEE",
    "C00742412": "BIDEN VICTORY FUND (JFC)",
    "C00742420": "BIDEN ACTION FUND (JFC)",
    "C00762591": "BIDEN/HARRIS 2024 VICTORY FUND",
    "C00725893": "DNC/BIDEN JOINT FUNDRAISING COMMITTEE",
    "C00703217": "DNC VICTORY FUND (JFC)",
    "C00401224": "DEMOCRATIC NATIONAL COMMITTEE"
}

new_rows_biden = []
updated_biden = 0

for cmte_id, cmte_name in BIDEN_COMMITTEES.items():
    mask = comm[cid_col].astype(str) == cmte_id

    if mask.any():
        # UPDATE existing rows
        comm.loc[mask, cand_col] = BIDEN_CAND_ID
        if party_col:  comm.loc[mask, party_col] = BIDEN_PARTY
        if desig_col:  comm.loc[mask, desig_col] = "Authorized by a candidate"
        if ctype_col:  comm.loc[mask, ctype_col] = "Presidential"
        updated_biden += mask.sum()
    else:
        # ADD new entries
        row = {
            cid_col: cmte_id,
            cand_col: BIDEN_CAND_ID,
        }
        if name_col:  row[name_col]  = cmte_name
        if party_col: row[party_col] = BIDEN_PARTY
        if desig_col: row[desig_col] = "Authorized by a candidate"
        if ctype_col: row[ctype_col] = "Presidential"
        if state_col: row[state_col] = "UNKNOWN"
        new_rows_biden.append(row)

if new_rows_biden:
    comm = pd.concat([comm, pd.DataFrame(new_rows_biden)], ignore_index=True)

print("Biden committees updated:", updated_biden)
print("Biden committees added:", len(new_rows_biden))
print("New committee_master shape:", comm.shape)

biden_comm = comm[comm[cand_col] == BIDEN_CAND_ID]
print(biden_comm[[cid_col, name_col, cand_col]].drop_duplicates().head(20))
```
