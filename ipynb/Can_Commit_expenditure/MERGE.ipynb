```python
import pandas as pd

# Load the three datasets
df_2019_2020 = pd.read_csv("Expen_Can_Com_Indep_2019_2020.csv")
df_2021_2022 = pd.read_csv("Expen_Can_Com_Indep_2021_2022.csv")
df_2023_2024 = pd.read_csv("Expen_Can_Com_Indep_2023_2024.csv")

# Concatenate all three datasets
df_combined = pd.concat([df_2019_2020, df_2021_2022, df_2023_2024], ignore_index=True)

```

```python
df_combined['TRANSACTION_DT'] = pd.to_datetime(df_combined['TRANSACTION_DT'], format='%m%d%Y')
```

```python
# Filter for date range: 07/01/2020 to 12/31/2024
start_date = pd.to_datetime('2020-07-01')
end_date = pd.to_datetime('2024-12-31')

df_combined = df_combined[(df_combined['TRANSACTION_DT'] >= start_date) & 
                          (df_combined['TRANSACTION_DT'] <= end_date)]
```

```python
# Save the combined dataset
df_combined.to_csv("Presidential_Campaign_Finance_2020_2024.csv", index=False)

print(f"Combined dataset shape: {df_combined.shape}")
print(f"Saved as: Presidential_Campaign_Finance_2020_2024.csv")
```
