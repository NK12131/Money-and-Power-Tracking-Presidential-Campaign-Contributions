```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
df1['CMTE_ID'].unique()
```

```python
df1['STATE'].unique()
```

```python
# Filter for US states only (50 states + DC)
us_states = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 
             'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 
             'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 
             'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 
             'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY', 'DC']

df1 = df1[df1['STATE'].isin(us_states)]
```

```python
df1.shape
```

```python
df1 = df1.drop(columns=['MEMO_CD', 'SUB_ID', 'ZIP_CODE', 'CITY','ZIP_CODE'])
df1.shape
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(subset=[
    'TRANSACTION_DT',
    'TRANSACTION_AMT',
    'STATE',
    'ENTITY_TP',
    'TRANSACTION_TP',
    'NAME'
])
print(df1.isna().sum())
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1["CMTE_ID"].unique()
```

```python
 # Check if CMTE_ID has any NaN values
df1["CMTE_ID"].isna().sum()
df1["CMTE_ID"].isin(["00", "00000000", "0"])

```

```python
print(df1["OCCUPATION"].unique())
```

```python
# Count values for OCCUPATION
df1['OCCUPATION'].value_counts()
```

```python
df1['EMPLOYER'] = df1['EMPLOYER'].fillna('Undisclosed')
df1['OCCUPATION'] = df1['OCCUPATION'].fillna('Undisclosed')
```

```python
df1['EMPLOYER'].value_counts()
```

```python
df1['OCCUPATION'].value_counts()
```

```python
df1.isnull().sum()
```

```python
df1["TRANSACTION_AMT"].unique()
```

```python
# Convert transaction amount to numeric
df1["TRANSACTION_AMT"] = pd.to_numeric(df1["TRANSACTION_AMT"], errors="coerce")

# Drop any NaN (invalid amounts)
df1 = df1.dropna(subset=["TRANSACTION_AMT"])

min_amt = df1["TRANSACTION_AMT"].min()
max_amt = df1["TRANSACTION_AMT"].max()

print(f"ðŸ’µ Transaction amount range: {min_amt:.0f}  â†’  {max_amt:.0f}")
```

```python
## import pandas as pd

# âœ… Keep only positive amounts (0 < amt â‰¤ 125000000)
df1 = df1[(df1["TRANSACTION_AMT"] > 0) & (df1["TRANSACTION_AMT"] <= 125_000_000)]

# Optional: reset index for cleanliness
df1 = df1.reset_index(drop=True)

# Verify range
min_amt = df1["TRANSACTION_AMT"].min()
max_amt = df1["TRANSACTION_AMT"].max()
print(f"ðŸ’µ Cleaned transaction amount range: {min_amt:.0f} â†’ {max_amt:.0f}")

```

```python
df1.shape
```

```python
df1.head(20)
```

```python
df1["ENTITY_TP"].unique()
```

```python
# Replace ENTITY_TP codes with full descriptive names
entity_map = {
    'IND': 'Individual',
    'ORG': 'Organization',
    'COM': 'Committee',
    'CCM': 'Candidate Committee',
    'CAN': 'Candidate',
    'PAC': 'Political Action Committee',
    'PTY': 'Party Organization'
}

df1['ENTITY_TP'] = df1['ENTITY_TP'].replace(entity_map)
```

```python
df1["ENTITY_TP"].unique()
```

```python
df1.shape
```

```python
df1.head()
```

```python
state_names = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas',
    'CA': 'California', 'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware',
    'FL': 'Florida', 'GA': 'Georgia', 'HI': 'Hawaii', 'ID': 'Idaho',
    'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada',
    'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York',
    'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah',
    'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia',
    'WI': 'Wisconsin', 'WY': 'Wyoming', 'DC': 'District of Columbia',
    'PR': 'Puerto Rico', 'GU': 'Guam', 'VI': 'U.S. Virgin Islands',
    'AS': 'American Samoa', 'MP': 'Northern Mariana Islands'
}

df1['STATE'] = df1['STATE'].replace(state_names)
```

```python
df1["STATE"].unique()
```

```python
df1["TRANSACTION_TP"].unique()
```

```python
# Keep only contributions (incoming money)
contribution_codes = ['10', '11', '15', '15C', '15E']
df1 = df1[df1['TRANSACTION_TP'].isin(contribution_codes)]
```

```python
df1.head()
```

```python
# Map TRANSACTION_TP codes to descriptive, readable names
transaction_map = {
    '10': '(10) Contribution to Candidate',
    '11': '(11) Contribution to Party Committee',
    '15': '(15) Contribution from Individual',
    '15C': '(15C) Candidate Self-Funding',
    '15E': '(15E) Earmarked Contribution'
}

df1['TRANSACTION_TP'] = df1['TRANSACTION_TP'].replace(transaction_map)
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1["TRANSACTION_DT"].unique()
```

```python
# Convert TRANSACTION_DT to string first, then to datetime
df1['TRANSACTION_DT'] = df1['TRANSACTION_DT'].astype(int).astype(str)
df1['TRANSACTION_DT'] = pd.to_datetime(df1['TRANSACTION_DT'], format='%m%d%Y')
```

```python
# Check if sorted in ascending order
is_sorted = df1["TRANSACTION_DT"].is_monotonic_increasing

print("âœ… Is TRANSACTION_DT sorted in ascending order?", is_sorted)
```

```python
df1.head()
```

```python
# Filter for dates between Jan 1, 2021 and Dec 31, 2022
df1 = df1[(df1['TRANSACTION_DT'] >= '2021-01-01') & (df1['TRANSACTION_DT'] <= '2022-12-31')]

# Verify the date range
print(df1['TRANSACTION_DT'].min())
print(df1['TRANSACTION_DT'].max())
```

```python
# Sort by TRANSACTION_DT in ascending order
df1 = df1.sort_values('TRANSACTION_DT')

# Verify it's sorted
print("âœ… Is TRANSACTION_DT sorted in ascending order?", df1["TRANSACTION_DT"].is_monotonic_increasing)
```

```python
df1.head(20)
```

```python
df1.tail(20)
```

```python
df1.shape
```

```python
# Save cleaned dataset for Tableau
df1.to_csv("Individual_contribution_2021_2022.csv", index=False)
print("âœ… File saved as: Individual_contribution_2021_2022.csv")
```
