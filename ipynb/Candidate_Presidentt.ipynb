```python
Keep_colms = ['candidate_id', 'name', 'candidate_election_year', 'incumbent_challenge_full', 'Votes', 'office_full','party_full','receipts','disbursements','candidate_status_full','load_date']
# Filter to only keep columns that exist in your dataframe
d2 = d2[[col for col in Keep_colms if col in d2.columns]]
```

```python
d2.shape
```

```python
d2.isnull().sum()
```

```python
# Replace nulls in both columns
d2['party_full'] = d2['party_full'].fillna('INDEPENDENT')
d2['incumbent_challenge_full'] = d2['incumbent_challenge_full'].fillna('Challenger')

# Verify no nulls remain
print(d1[['party_full', 'incumbent_challenge_full']].isna().sum())
```

```python
d2.isnull().sum()
```

# MERGE

```python
# Concatenate d1 (2020) and d2 (2021-2024) vertically
combined_df = pd.concat([d1, d2], ignore_index=True)

# Verify the shape
print(combined_df.shape)  # Should be (1782, 8) = 177 + 1605 rows

# Check the years are in order
print(combined_df['candidate_election_year'].value_counts().sort_index())
```

```python
# Convert load_date to datetime
combined_df['load_date'] = pd.to_datetime(combined_df['load_date'])

# Verify the change
print(combined_df.dtypes)
```

```python
combined_df.to_csv('Candidate_Master.csv', index=False)
```
